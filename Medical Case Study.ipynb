```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('insurance.csv')

# Question 1: How is age related to medical costs?
print("Question 1: How is age related to medical costs?")
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='age', y='charges', alpha=0.6)
plt.title('Relationship Between Age and Medical Charges')
plt.xlabel('Age')
plt.ylabel('Charges ($)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# Question 2: How is number of children related to medical costs?
print("Question 2: How is number of children related to medical costs?")
child_charges = df.groupby('children')['charges'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(data=child_charges, x='children', y='charges', hue='children', palette='viridis', legend=False)
plt.title('Average Medical Charges by Number of Children')
plt.xlabel('Number of Children')
plt.ylabel('Average Charges ($)')
plt.show()

#Question 3: How is the number of people distributed by region?
print("Question 3: How is the number of people distributed by region?")
plt.figure(figsize=(10, 6))
region_counts = df['region'].value_counts().reset_index()
region_counts.columns = ['region', 'count']

sns.barplot(data=region_counts, x='region', y='count', hue='region', palette='muted', legend=False)
plt.title('Distribution of People by Region')
plt.xlabel('Region')
plt.ylabel('Count')
plt.show()

# Question 4: How is the number of people distributed by age?
print("Question 4: How is the number of people distributed by age?")
plt.figure(figsize=(10, 6))
sns.histplot(df['age'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of People by Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Question 5: How are the charges distributed?
print("Question 5: How are the charges distributed?")
plt.figure(figsize=(10, 6))
sns.histplot(df['charges'], bins=30, kde=True, color='salmon')
plt.title('Distribution of Medical Charges')
plt.xlabel('Charges ($)')
plt.ylabel('Frequency')
plt.show()
```
